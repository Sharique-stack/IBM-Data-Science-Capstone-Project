## Using <font color=blue> BeautifulSoup</font> for web scraping from wikipedia  

```python
import requests
import urllib.request

import time
import pandas as pd
from bs4 import BeautifulSoup
import io
```

fetching url data from web

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
req=requests.get(url)
```

```python
req
```

```python
soup=BeautifulSoup(req.content,'lxml')
```

```python
table = soup.find_all('table')[0]

df = pd.read_html(str(table))


neighborhood=pd.DataFrame(df[0])
```

```python
neighborhood
```

# Cleaning

Removing rows with 'Not assigned'

```python
neighborhood=neighborhood[neighborhood['Borough'] != 'Not assigned']
```

```python
neighborhood.set_index('Postal Code',inplace=True)
neighborhood
```

```python
import geocoder # import geocoder

# looping through the postal_codes
    # initialize your variable to None
lat_lng_coords = None
postal_code=neighborhood['Postal Code'].iloc[0]

# loop until you get the coordinates
while(lat_lng_coords is None):
    g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
    lat_lng_coords = g.latlng

latitude = lat_lng_coords[0]
longitude = lat_lng_coords[1]
(neighborhood['Latitude'])[0]=latitude
(neighborhood['Longitude'])[0]=longitude
```

```python
url="https://raw.githubusercontent.com/Sharique-stack/IBM-Data-Science-Capstone-Project/master/Geospatial_Coordinates.csv"
s=requests.get(url).content
latlong=pd.read_csv(io.StringIO(s.decode('utf-8')))
latlong=pd.DataFrame(latlong)
latlong.set_index('Postal Code', inplace=True)
latlong
```

```python
neighborhood['Latitude']=''
neighborhood['Longitude']=''
for i in neighborhood.index:
    neighborhood.loc[i,'Latitude']=latlong.loc[i,'Latitude']
    neighborhood.loc[i,'Longitude']=latlong.loc[i,'Longitude']    
```
